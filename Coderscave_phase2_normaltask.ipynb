## **Customer Sentiment Analysis and Feedback Insights**

```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
import pandas as pd
```

```python
# Download VADER lexicon
nltk.download('vader_lexicon')
```

```python
# Read your dataset (replace 'your_dataset.csv' with the actual file path or URL)
df = pd.read_csv('/content/purchase behaviou dataset.csv')
df
```

```python
df.head(5)
```

```python
df.describe()
```

```python
# Check if the necessary columns are present in the dataset
required_columns = ['LYLTY_CARD_NBR', 'LIFESTAGE', 'PREMIUM_CUSTOMER']
for col in required_columns:
    if col not in df.columns:
        raise KeyError(f"The required column '{col}' is not present in the dataset.")
```

```python
# Create a synthetic 'Review' column by combining existing columns
df['Review'] = df['LYLTY_CARD_NBR'].astype(str) + ' ' + df['LIFESTAGE'] + ' ' + df['PREMIUM_CUSTOMER']
```

```python
# Initialize the VADER sentiment analyzer
sia = SentimentIntensityAnalyzer()
sia
```

```python
# Function to get sentiment analysis results
def get_sentiment(text):
    score = sia.polarity_scores(text)['compound']
    if score >= 0.05:
        return 'Positive'
    elif score <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'
```

```python
# Apply sentiment analysis to the reviews in the dataset
df['Sentiment'] = df['Review'].apply(get_sentiment)
```

```python
# Display the results
print("\nSentiment Analysis Results:")
print(df[['LYLTY_CARD_NBR', 'LIFESTAGE', 'PREMIUM_CUSTOMER', 'Review', 'Sentiment']])
```
